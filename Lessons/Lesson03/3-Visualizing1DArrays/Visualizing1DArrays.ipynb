## Single 1D arrays

A single 1D array can be generated and visualized as follows:


```python
import numpy as np
from matplotlib import pyplot as plt

x = np.array( [2, 0, 5, 1, 8] )

plt.figure()
plt.plot(x)
plt.show()
```

Note that the horizontal axis represents array position, and the vertical axis represents the value of the array element.

Let's repeat with a longer array, one that has 100 elements:

```python
np.random.seed(0)         # set the random number generator state
x = np.random.randn(100)  # 100 random numbers from the standard normal distribution

plt.figure()
plt.plot(x)
plt.show()
```

## Multiple 1D arrays

A 2D NumPy array can be regarded as a collection of 1D arrays.  For example, the following array is 2 x 5, and this can be regarded as two 5-element 1D arrays (or five 2-element 1D arrays).

To plot multiple 1D arrays that are saved in a 2D array requires appropriate array [transposing](https://en.wikipedia.org/wiki/Transpose).

```python
a = np.array( [ [2, 0, 5, 1, 8] , [12, 10, 15, 11, 18] ] )

print(a)
print()
print( a.shape )

plt.figure()
plt.plot(a)
plt.show()
```

Note that `plt.plot` has produced one line for each column of `a`.

We can instead plot one line for each row of `a` by using `.T` to transpose the array, like this:

```python
plt.figure()
plt.plot(a.T)
plt.show()
```

## Plotting two single 1D arrays

It is often desirable to check how two or more 1D arrays relate to each other.  This can be achieved using `plt.plot(x,y)`, where `x` and `y` are arrays for the horizontal and vertical axes, respectively.  For example: 

```python
x = np.array([0, 1, 5, 6, 10])
y = np.array([10, 5, 8, 7, 8])


plt.figure()
plt.plot(x, y)
plt.show()
```

This also works when `y` is a 2D array (i.e., a collection of 1D arrays).

```python
x = np.array([0, 1, 5, 6, 10])
y = np.array( [ [2, 0, 5, 1, 8] , [12, 10, 15, 11, 18] ] )


plt.figure()
plt.plot(x, y.T)
plt.show()
```

Sometimes `x` does not contain uniformly increasing values, like this:

```python
x = np.array([0, 10, 6, 1, 5])
y = np.array( [ [2, 0, 5, 1, 8] , [12, 10, 15, 11, 18] ] )


plt.figure()
plt.plot(x, y.T)
plt.show()
```

In cases like this, it is sometimes better to plot symbols rather than lines. One way to do this is to use circle symbols like this.

```python
plt.figure()
plt.plot(x, y.T, 'o')
plt.show()
```

For more details regarding different ways to use `plt.plot` refer to the relevant [Matplotlib documentation](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.plot.html).
