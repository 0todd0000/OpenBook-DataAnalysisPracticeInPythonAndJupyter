2D arrays usually contain hundreds or thousands of values, and our brains our incapable of understanding the 2D patterns in so many numbers. Matplotlib offers many functions for plotting 2D data, and two key functions are `plt.pcolor` and `plt.imshow`.

___

## plt.pcolor

`pcolor` means "psuedo-color", which in turn means that numbers are assigned colors. Here is a simple example:


```python
import numpy as np
from matplotlib import pyplot as plt

a = np.array([  [1,2,3], [4,5,6] ])

plt.figure()
plt.pcolor(a)
plt.colorbar()
plt.show()
```

Note that the `pcolor` command produces a pseudo-color plot and that `colorbar` generates the color bar on the right. Colors represent values.

Note that the first row (with values: `[1,2,3]`) is on the **bottom** of the pcolor plot.

We can change various aspects of the `pcolor` plot including the [colormap](https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html):

```python
plt.figure()
plt.pcolor(a, cmap='jet')
plt.colorbar()
plt.show()
```

The value of `pcolor` becomes more obvious when dealing with larger arrays, and arrays that have spatial structure, like the example below. Note that this example is a modified version of [this example](https://matplotlib.org/3.3.0/gallery/images_contours_and_fields/pcolor_demo.html) from Matplotlib's documentation.

```python
n   = 100
x,y = np.meshgrid(np.linspace(-3, 3, n), np.linspace(-2, 2, n))
z   = np.exp(-x**2 - y**2)

plt.figure()
plt.pcolor(x, y, z, cmap='PuBu_r')
plt.show()
```

In the example above, `linspace` generates `n` equally spaced values between the first and second arguments (e.g., between -3 and 3):

```python
print( np.linspace(-3, 3, n) )
```

The `meshgrid` function generates a 2D array of $x$ and $y$ coordinates. Let's look at just the first five rows and columns of both `x` and `y`:

```python
x,y = np.meshgrid(np.linspace(-3, 3, n), np.linspace(-2, 2, n))

print( x[:5,:5] )
print()
print( y[:5,:5] )
```

Note that the `x` values increase from left to right, and that the `y` values increase from top to bottom.

The $z$ variable is the variable that is actually plotted using `pcolor`, and it is a function of $x$ and $y$:

<center>
    $z(x,y) = e^{(-x^2 - y^2)}$
</center>

Refer to Matplotlib's [pcolor documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.pcolor.html) for more `pcolor` details and options.

## plt.imshow

An alternative to `pcolor` is `imshow` or "image show". Let's see how it works for the examples above:

```python
a = np.array([  [1,2,3], [4,5,6] ])

plt.figure()
plt.imshow(a, cmap='jet')
plt.colorbar()
plt.show()
```

Note that this is nearly identical to `pcolor`, except that the first row is on the top. This is the normal convention for images: y values increase from top to bottom.

```python
n   = 100
x,y = np.meshgrid(np.linspace(-3, 3, n), np.linspace(-2, 2, n))
z   = np.exp(-x**2 - y**2)

plt.figure()
plt.imshow(z, cmap='PuBu_r')
plt.show()
```

This, again, is almost the same as the `pcolor` result. A key difference is that, whereas `pcolor` can accept **three** input array arguments (x, y, z), `imshow` can only accept one input array argument. Thus `imshow`, like most image display software, assumes that the values are equally spaced in the $x$ and $y$ directions. `pcolor` more flexibly allows arbitrary $x$ and $y$ spacing.

A key advantage of `imshow` is that it can be used to display images, like this:

```python
fname = 'img/1206_FMRI.jpg'
I     = plt.imread(fname)

plt.figure()
plt.imshow(I)
plt.show()
```

Note that the image file `1206_FMRI.jpg` is saved in a directory called **img**, and that the **img** directory is in the same directory as this notebook.
