Scatter plots and bar plots are very common in science, and they can both be easily produced using Matplotlib.

___

## Scatter plots

Here is an example scatter plot:


```python
import numpy as np
from matplotlib import pyplot as plt


# Create random data:
np.random.seed(0)
x = np.random.randn(50)
y = np.random.randn(50)


# Create figure:
plt.figure()
plt.scatter(x, y, color='r', alpha=0.3, s=400)
plt.show()
```

Here the parameters `alpha` and `s` represent transparency and marker size, respectively.

For more details, see Matplotlib's [scatterplot documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html).

___

## Bar plots

Bar plots can also be easily generated, like this:

```python
x = [1, 2, 5, 6]
y = [10, 12,  20, 16]


# Create figure:
plt.figure()
plt.bar(x, y)
plt.show()
```

See more details and examples in Matplotlib's [bar documentation](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.bar.html).

You may also be interested in Matplotlib's [errorbar function](https://matplotlib.org/1.2.1/examples/pylab_examples/errorbar_demo.html), which allows you to plot error bars on lines and or bars, like this:

```python
x = [1, 2, 5, 6]
y = [10, 12,  20, 16]
e = [3, 4, 5, 2]

# Create figure:
plt.figure()
plt.bar(x, y)
plt.errorbar(x, y, yerr=e, fmt='o')
plt.show()
```
