There are many different ways to customize and/or add details to Matplotlib figures. This notebook reviews a few key options.  For more comprehensive possibilities, refer to Matplotlib's [example gallery](https://matplotlib.org/3.1.1/gallery/index.html).

___

## Axis labels

Axis labels can be added using the `xlabel` and `ylabel` commands, which also allow you to set the font, size, color, etc.


```python
import numpy as np
from matplotlib import pyplot as plt

a = np.array([1,3,2,8,5])

plt.figure()
plt.plot(a)
plt.xlabel('Time (s)', size=16)
plt.ylabel('Position (m)', color='r', name='Times New Roman')
plt.show()
```

___

## Axis tick labels

You can substitute arbitrary labels for the axis ticks using the `xticks` or `yticks` command like this:


```python
x = [0, 1, 2, 3, 4]
a = np.array([1,3,2,8,5])
xlabels = ['Label 0', 'Hello', 'I', 'Love', 'Matplotlib']

plt.figure()
plt.plot(x, a)
plt.xticks( ticks=x, labels=xlabels , size=9, color='blue' )
plt.show()
```

___

## Axis titles

A title for the axis can be specified using the `title` function like this:

```python
plt.figure()
plt.plot(x, a)
plt.title('This is an excellent figure')
plt.show()
```

___

## Legends

When multiple lines or other types of plots are present in one figure, it is often useful to add a legend which explains the differences between those plot elements. This can be done using the `legend` command, like this:

```python
x0     = [0, 1, 2, 3, 4]
x1     = [1, 3, 2, 8, 5]
labels = ['Before', 'After']

plt.figure()
plt.plot(x0)
plt.plot(x1)
plt.legend( labels )
plt.show()
```

You can achieve the same thing by using the `label` keyword when creating each plotted element, like this:

```python
plt.figure()
plt.plot(x0, label='Before')
plt.plot(x1, label='After')
plt.legend()
plt.show()
```

___

## Line properties

A variety of properties can be adjusted including line width, color, marker characteristics, etc.  Here is one example:

```python
x0 = [0, 1, 2, 3, 4]
x1 = [1, 3, 2, 8, 5]

plt.figure()
plt.plot(x0, linewidth=5, color='b', marker='^', markersize=10, markerfacecolor='white')
plt.plot(x1, lw=2, color='r', marker='o', ms=16, mfc=(1,0.9,0.9))
plt.show()
```

Note that the following keyword arguments are equivalent:

* `linewidth` and `lw`
* `markersize` and `ms`
* `markerfacecolor` and `mfc`

Note also that colors can be speficied using either [color names](https://matplotlib.org/3.1.0/gallery/color/named_colors.html) or using a 3-tuple representing [RGB colors](https://en.wikipedia.org/wiki/RGB_color_model), where 1 is the maximum value.
