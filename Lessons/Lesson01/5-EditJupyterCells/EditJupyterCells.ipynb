<img src="img/python-2.png" alt="ascreenshot" width=600/>

# Markdown cells
[Back to Table of Contents](#toc)
<br>
<br>

* Select **Insert** .. **Insert Cell Below**  &nbsp; &nbsp;(or Insert Cell Above) &nbsp; &nbsp; (same as [above](#python))


* Change the cell type to **Markdown**.


<img src="img/markdown-1.png" alt="ascreenshot" width=600/>

<br>
<br>

* Enter Markdown code, like `# This is a Heading`
* Select **Run** to display the Markdown code as HTML.

<img src="img/markdown-2.png" alt="ascreenshot" width=600/>

<br>
<br>


* Verify that the displayed HTML looks correct.
