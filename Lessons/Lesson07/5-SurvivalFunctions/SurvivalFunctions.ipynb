Analytical survival functions can be plotted (similar to the PDFs in Fig.7.1) as follows:

<br>

```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

```python
# Define ranges of thresholds
u0   = np.linspace(-1, 2, 1001)  # random variable range for the uniform distribution
u1   = np.linspace(-4, 4, 1001)  # random variable range for the normal distribution
u2   = np.linspace(0, 20, 1001)  # random variable range for the chi2 distribution


# Use analytical equations to calculate the probability densities at all x values 
sf0  = stats.uniform.sf(u0)     # SF for a uniform random variable
sf1  = stats.norm.sf(u1)        # SF for a normal random variable
sf2  = stats.chi2.sf(u2, 4)     # SF for a chi2 random variable (with degrees of freedom = 4)


# Plot the survival functions
fig,ax = plt.subplots(1, 3, figsize=(14,3))

ax[0].plot(u0, sf0)
ax[1].plot(u1, sf1)
ax[2].plot(u2, sf2)

ax[0].set_title('Uniform')
ax[1].set_title('Normal')
ax[2].set_title('$\chi^2$  ($v$ = 4)')

plt.setp(ax, xlabel='u')
plt.setp(ax, ylabel='Survival function')

[axx.axhline(0, color='k', linestyle=':')  for axx in ax]
[axx.axhline(1, color='k', linestyle=':')  for axx in ax]

plt.show()
```

<b>Figure 7.2</b>:  Survival functions for three types of random variables. $u$ = threshold. $\nu$ = degrees of freedom.

<br>
<br>

From these plots the probability $P(x>u)$ is a somewhat clearer. For example, it is much clearer that $P(x>2.0)$ is close to zero for the normal distribution.

## Numerically verifying survival functions

Analytical survival functions can be numerically verified in Python as demonstrated below.

**Note**:  only the unform survival function is verified below. As an exercise, you may wish to numerically verify the survival functions for the normal and $\chi^2$ distributions.

<br>

```python
x0   = np.random.rand(10000)   # random values from the uniform distribution
u0n  = np.linspace(-1, 2, 51)  # thresholds for numerical survival function calculations
sf0n = [(x0>uu).mean()  for uu in u0n]  # numerical survival function calculations


# plot the analytical and numerical survival functions:
plt.figure()
ax = plt.axes()
ax.plot(u0, sf0, linewidth=2, zorder=1, label='Analytical')
ax.plot(u0n, sf0n, 'o', zorder=0, label='Numerical')
ax.axhline(0, color='k', linestyle=':')
ax.axhline(1, color='k', linestyle=':')
ax.legend()
ax.set_xlabel('u')
ax.set_ylabel('P(x>u)')
plt.show()
```

<b>Figure 7.3</b>:  Numerical verification of the survival function for the uniform distribution.

<br>
<br>


Note that the numerical survival function is nearly identical to the analytical one.
